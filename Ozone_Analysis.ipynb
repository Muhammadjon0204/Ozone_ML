```python
import pandas as pd
import matplotlib.pyplot as plt
import sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer

# Загружаем данные и сразу убираем предупреждения о типах
data = pd.read_csv('air_data.csv',low_memory=False)

# Эта команда выведет первые 5 строк твоей таблицы прямо в ноутбук
data.head()
```

```python
# 1. Выбираем целевую переменную (y) и признаки (X)
y = data['Arithmetic Mean']
features = ['Latitude', 'Longitude', 'AQI', 'Observation Count', 'Observation Percent']
X = data[features]

# 2. Разделяем на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=42)

# 3. Заполняем пропуски (Imputation)
imputer = SimpleImputer(strategy='mean')
imputed_X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
imputed_X_test = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)

print("Данные успешно подготовлены и очищены!")
```

```python
# 1. Обучаем модель (Random Forest)
print("Начинаю обучение модели на 300к+ строках...")
model = RandomForestRegressor(n_estimators=50, random_state=1, n_jobs=-1)
model.fit(imputed_X_train, y_train)

# 2. Делаем предсказание и считаем ошибку
preds = model.predict(imputed_X_test)
mae = mean_absolute_error(y_test, preds)

print("-" * 30)
print(f"Средняя ошибка (MAE): {mae:.6f}")
print("-" * 30)

# 3. Строим график сравнения (первые 100 точек)
plt.figure(figsize=(14, 7))
plt.plot(y_test.values[:100], label='Реальные данные (EPA)', color='#1f77b4', linewidth=2, alpha=0.6)
plt.plot(preds[:100], label='Предсказание модели', color='#d62728', linestyle='--', linewidth=2)

plt.title('Анализ точности ИИ: Реальность vs Предсказание', fontsize=14)
plt.xlabel('Номер замера', fontsize=12)
plt.ylabel('Концентрация Озона', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# это для себя сделал, было интересно
import seaborn as sns

# Смотрим, какие данные были самыми важными для предсказания
importances = model.feature_importances_
feature_names = features
feature_importance_df = pd.DataFrame({'Признак': feature_names, 'Важность': importances})
feature_importance_df = feature_importance_df.sort_values(by='Важность', ascending=False)

# Строим график важности
plt.figure(figsize=(10, 5))
sns.barplot(x='Важность', y='Признак', data=feature_importance_df, palette='viridis')
plt.title('Какие факторы больше всего влияют на уровень озона?')
plt.show()
```

```python
# наглядный пример насколько эффективно будет предсказывать модель на конкретном примере
# опять же это было длч себя

# Выбираем случайный пример из тестовых данных (например, под индексом 10)

index = 177

# Получаем реальное значение и то, что предсказал наша модель
real_value = y_test.values[index]
predicted_value = preds[index]

print(f"--- Проверка на конкретном примере (индекс {index}) ---")
print(f"Реальная концентрация озона: {real_value:.6f}")
print(f"Предсказание модели:          {predicted_value:.6f}")
print(f"Разница (ошибка):             {abs(real_value - predicted_value):.6f}")

# Для наглядности: во сколько раз ошибка меньше реального значения
accuracy_pct = (1 - abs(real_value - predicted_value) / real_value) * 100
print(f"Точность на этом примере:    {accuracy_pct:.2f}%")
```
